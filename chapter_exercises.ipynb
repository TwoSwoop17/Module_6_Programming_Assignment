```python
import datetime

# Get the current date/time and convert it to a string
now = datetime.datetime.now()
today_str = str(now)

# Write it to today.txt
with open('today.txt', 'w') as f:
    f.write(today_str)

print(today_str)
```

```python
with open('today.txt', 'r') as f:
    today_string = f.read()

print(today_string)
```

```python
from dateutil import parser

parsed_date = parser.parse(today_string)
print(parsed_date)
print(type(parsed_date))
```

```python
%%writefile worker.py
import time
import random

def sleep_and_print():
    """Sleep a random amount of time between 0 and 1 seconds, then print the current time."""
    time.sleep(random.random())
    print(time.ctime())
```

```python
import multiprocessing
from worker import sleep_and_print

if __name__ == '__main__':
    processes = [multiprocessing.Process(target=sleep_and_print) for _ in range(3)]

    for p in processes:
        p.start()

    for p in processes:
        p.join()

    print('Done!')
```
