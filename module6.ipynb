```python
import datetime


now = datetime.datetime.now()
today_str = str(now)

with open('today.txt', 'w') as f:
    f.write(today_str)

print(today_str)
```

```python
with open('today.txt', 'r') as f:
    today_string = f.read()

print(today_string)
```

```python
from dateutil import parser

parsed_date = parser.parse(today_string)
print(parsed_date)
print(type(parsed_date))
```

```python
%%writefile worker.py
import time
import random
import os

def sleep_and_print(q):
    """Wait a random 0-1 seconds, then report the current time (and this process's PID)
    back to the parent through the queue, and exit."""
    time.sleep(random.random())
    q.put(f"PID {os.getpid()} woke up at {time.ctime()}")
```

```python
import multiprocessing
from worker import sleep_and_print

if __name__ == '__main__':
    q = multiprocessing.Queue()

    processes = [
        multiprocessing.Process(target=sleep_and_print, args=(q,))
        for _ in range(3)
    ]

    
    for p in processes:
        p.start()
    for p in processes:
        p.join()

    
    while not q.empty():
        print(q.get())

    print('Done!')
```
